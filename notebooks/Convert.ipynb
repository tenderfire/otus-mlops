```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession
        .builder
        .appName("OTUS")
        .getOrCreate()
)
```

```python
riiid_df = spark.read.csv("riiid/train.csv", inferSchema=True, header=True)
```

Поделим данные на 10 разделов и сохраним в формате parquete

```python
%%time
riiid_df.count()
```

```python
(
    riiid_df
        .repartition(10)
        .write
        .mode("overwrite")
        .parquet("riiid/train.parquet")
)
```
